### Dataset q1


```python
income = pd.read_csv("/Users/jaredmcmullen/Desktop/GSB-S545/data/income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
life_expectancy = pd.read_csv("/Users/jaredmcmullen/Desktop/GSB-S545/data/life_expectancy_years.csv")
population = pd.read_csv("/Users/jaredmcmullen/Desktop/GSB-S545/data/population_total.csv")    
four_regions = pd.read_csv("/Users/jaredmcmullen/Desktop/GSB-S545/data/four_regions.csv")
```

```python
# Unpivot the datasets
df_income = income.melt(id_vars = "country", var_name = "Year", value_name = "income")
df_life_expectancy = life_expectancy.melt(id_vars = "country", var_name = "Year", value_name = "life_exp")
df_population = population.melt(id_vars = "country", var_name = "Year", value_name = "population")

# Filter datasets for only year 2010
df_income = df_income[df_income.Year == "2010"]
df_life_expectancy = df_life_expectancy[df_life_expectancy.Year == "2010"]
df_population = df_population[df_population.Year == "2010"]

#replacing values
df_population['population'] = df_population['population'].replace('\.', "", regex = True)
df_population['population'] = df_population['population'].replace('M', "000000", regex = True)
df_population['population'] = df_population['population'].replace('B', "000000000", regex = True)
df_population['population'] = df_population['population'].replace('k', "000", regex = True)
df_income['income'] = df_income['income'].replace('k', "000", regex = True)
df_income['income'] = df_income['income'].replace('\.', "", regex = True)

#converting values
df_population['population'] = df_population['population'].astype(int)
df_income['income'] = df_income['income'].astype(int)

# Merge the datasets
df_combined_initial = df_population.merge(df_income, on = ["country", "Year"]).merge(df_life_expectancy, on = ["country", "Year"])
df_combined = df_combined_initial.merge(four_regions, on = ["country"])
df_combined
```
